```python
## Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
from math import log
from scipy.special import boxcox1p
from scipy import stats
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis



import pickle
import requests
import json

```

```python
#### Read dataset 
data = pd.read_csv("data/heart.csv")
data.head()
```

```python
org_data = data
```

```python
## Dummy variables
sex = pd.get_dummies(org_data['sex'])
sex.columns = ["female", "male"]
cp = pd.get_dummies(org_data['cp'])
cp.columns = ["Typical Angina", "Atypical Angina", "Non Anginal Pain", "Asymptomatic"]
fbs = pd.get_dummies(org_data['fbs'])
fbs.columns = ["false", "true"]
restecg = pd.get_dummies(org_data['restecg'])
restecg.columns = ["Normal", "Having ST-T Wave Abnormality", "Showing Probable or Definite Left Ventricular Hypertrophy"]
exang = pd.get_dummies(org_data['exang'])
exang.columns = ["no", "yes"]
slope = pd.get_dummies(org_data['slope'])
slope.columns = ["Upsloping", "Flat", "Downsloping"]
ca = pd.get_dummies(org_data['ca'])
ca.columns = ['ca0', 'ca1', 'ca2', 'ca3', 'ca4']
thal = pd.get_dummies(org_data['thal'])
thal.columns = ["Unknown", "thal Normal", "Fixed Defect", "Reversible Defect"]
```

```python
## Append dummy variables to quantitative variable data 
org_data_target = org_data["target"]
org_data_quantitative = org_data.drop(columns = ["sex", "cp", "fbs", "restecg", "exang", "slope", "ca", "thal", "target"])
org_data_all = org_data_quantitative.join(sex).join(cp).join(fbs).join(restecg).join(exang).join(slope).join(ca).join(thal)
```

### Logistic Regression

```python
clf = LogisticRegression(random_state=0, solver='newton-cg', multi_class='multinomial').fit(X_train, y_train)
train_score = clf.score(X_train, y_train)
test_score = clf.score(X_test, y_test)
print(f"Logistic Regression Method: Train Score: {train_score:.3f}, Testing Score: {test_score:.3f}")
```

```python
## Confusion matrix for testing set
row_1 = confusion_matrix(y_test, clf.predict(X_test))[0,]
row_2 = confusion_matrix(y_test, clf.predict(X_test))[1,]

table = pd.DataFrame([row_1, row_2])
table.columns = ["Observed Negative", "Observed Positive"]
table.index = ["Predicted Negative", "Predicted Positive"]
table
```

```python
## Classification report for testing set
print(classification_report(y_test, clf.predict(X_test)))
```

### KNN Method

```python
# Loop through different k values to see which has the highest accuracy
# Note: We only use odd numbers because we don't want any ties
train_scores = []
test_scores = []
for k in range(1, 20, 2):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_scaled, y_train)
    train_score = knn.score(X_train_scaled, y_train)
    test_score = knn.score(X_test_scaled, y_test)
    train_scores.append(train_score)
    test_scores.append(test_score)
    print(f"k: {k}, Train/Test Score: {train_score:.3f}/{test_score:.3f}")
    
    
plt.plot(range(1, 20, 2), train_scores, marker='o', label='training')
plt.plot(range(1, 20, 2), test_scores, marker="x", label='testing')
plt.legend()
plt.xlabel("k neighbors")
plt.ylabel("Testing accuracy Score")
plt.show()
```

### LDA Method

```python
clf = LinearDiscriminantAnalysis()
clf.fit(X_train, y_train)  
train_score = clf.score(X_train, y_train)
test_score = clf.score(X_test, y_test)
print(f"LAD Method: Train Score: {train_score:.3f}, Testing Score: {test_score:.3f}")
```

### QDA Method

```python
clf = QuadraticDiscriminantAnalysis()
clf.fit(X_train, y_train)  
train_score = clf.score(X_train, y_train)
test_score = clf.score(X_test, y_test)
print(f"QDA Method: Train Score: {train_score:.3f}, Testing Score: {test_score:.3f}")
```
